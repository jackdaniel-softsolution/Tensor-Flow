```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
df["income_bracket"]=le.fit_transform(df["income_bracket"])
```

```python
df.head()
```

## Train test split

```python
x_data = df.drop('income_bracket',axis=1)
y_labels = df['income_bracket']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data,y_labels,test_size=0.3,random_state=101)
```

```python
import tensorflow as tf
```

```python
df.columns
```

```python
g=tf.feature_column.categorical_column_with_vocabulary_list("gender",["Female","Male"])
ed=tf.feature_column.categorical_column_with_hash_bucket("education", hash_bucket_size=1000)
ms=tf.feature_column.categorical_column_with_hash_bucket("marital_status", hash_bucket_size=1000)
o=tf.feature_column.categorical_column_with_hash_bucket("occupation", hash_bucket_size=1000)
r=tf.feature_column.categorical_column_with_hash_bucket("relationship", hash_bucket_size=1000)
wc=tf.feature_column.categorical_column_with_hash_bucket("workclass", hash_bucket_size=1000)
nc=tf.feature_column.categorical_column_with_hash_bucket("native_country", hash_bucket_size=1000)
```

```python
age=tf.feature_column.numeric_column("age")
ed_num=tf.feature_column.numeric_column("education_num")
cg=tf.feature_column.numeric_column("capital_gain")
cl=tf.feature_column.numeric_column("capital_loss")
hpw=tf.feature_column.numeric_column("hours_per_week")
```

## features columns

```python
feat_cols=[g,ed,ms,o,r,wc,nc,age,ed_num,cg,cl,hpw]
```

## Estimator

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_cols)
```

## Input funtion

```python
input_func=tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=100,num_epochs=None,shuffle=True)
```

## Train the model

```python
model.train(input_fn=input_func,steps=5000)
```

## Prediction

```python
pred_input=tf.estimator.inputs.pandas_input_fn(x=x_test,batch_size=100,num_epochs=1,shuffle=False)
```

```python
predictions = list(model.predict(input_fn=pred_input))
```

```python
predictions[0]
```

```python
original_pred=[]
for i in predictions:
    original_pred.append(i['class_ids'][0])
    
    
```

```python
original_pred
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,original_pred))
```
